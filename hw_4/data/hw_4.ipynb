```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
file_path = "resultsGA.csv"
data = pd.read_csv(file_path)

# Display the first few rows
data.head()
```

```python
# Map categorical values to numerical values
data['crossover_type'] = data['crossover_type'].map({'ONE_POINT': 1, 'TWO_POINT': 2})

# Display the first few rows to verify the changes
data.head()
```

```python
# Check dataset information
data.info()

# Check for missing values
data.isnull().sum()

# Summary statistics
data[['measured_time_milliseconds', 'average_error']].describe()
```

```python
# Plot distribution of runtime
sns.histplot(data['measured_time_milliseconds'], kde=True)
plt.title('Runtime Distribution')
plt.xlabel('Time (ms)')
plt.ylabel('Frequency')
plt.savefig('data/runtime_dist.png')
plt.show()

# Plot distribution of average error
sns.histplot(data['average_error'], kde=True)
plt.title('Error Distribution')
plt.xlabel('Average Error')
plt.ylabel('Frequency')
plt.savefig('data/error_dist.png')
plt.show()
```

```python
sns.boxplot(x='n', y='measured_time_milliseconds', data=data)
plt.ylim(0, data['measured_time_milliseconds'].quantile(0.99))  # Remove one big outlier to see the boxplot better
plt.title('Impact of Instance size on Runtime')
plt.savefig('data/instance_size_runtime.png')
plt.show()

sns.boxplot(x='n', y='average_error', data=data)
plt.title('Impact of Instance size on Average Error')
plt.savefig('data/instance_size_error.png')
plt.show()
```

```python
# Correlation between numeric columns
correlation_matrix = data.corr()

# Set the figure size
plt.figure(figsize=(14, 10))

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 10})

# Rotate the x and y axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.title('Correlation Matrix')
plt.savefig('data/correlation_matrix.png', bbox_inches='tight')
plt.show()
```

```python
optimal = data.loc[data['average_error'].idxmin()]  # Find row with the smallest error
print("Optimal Parameters (Min Error):\n", optimal)

fastest = data.loc[data['measured_time_milliseconds'].idxmin()]  # Find row with the shortest runtime
print("Optimal Parameters (Min Runtime):\n", fastest)

worst_optimal = data.loc[data['average_error'].idxmax()]  # Find row with the biggest error
print("Worst Parameters (Biggest Error):\n", worst_optimal)

slowest = data.loc[data['measured_time_milliseconds'].idxmax()]  # Find row with the longest runtime
print("Worst Parameters (Biggest Runtime):\n", slowest)
```

# Best Parameters

```python
# Top 10 rows with minimal error
best_error = data.nsmallest(20, 'average_error')

# Bottom 10 rows with the highest error
worst_error = data.nlargest(20, 'average_error')

# Display
print("Top 10 by Error:\n", best_error)
print("Bottom 10 by Error:\n", worst_error)
```

```python
# Top 10 rows with minimal runtime
best_runtime = data.nsmallest(20, 'measured_time_milliseconds')

# Bottom 10 rows with the highest runtime
worst_runtime = data.nlargest(20, 'measured_time_milliseconds')

# Display
print("Top 10 by Runtime:\n", best_runtime)
print("Bottom 10 by Runtime:\n", worst_runtime)
```

```python
# Most common parameter values for the top 10 by error
print("Parameter Trends for Best Error:")
print(best_error[['n','population_size', 'mutation_rate', 'crossover_rate', 'tournament_size','num_elites','crossover_type', 'num_generations']].value_counts())

# Most common parameter values for the top 10 by runtime
print("Parameter Trends for Best Runtime:")
print(best_runtime[['n','population_size', 'mutation_rate', 'crossover_rate', 'tournament_size','num_elites','crossover_type', 'num_generations']].value_counts())

# Most common parameter values for the top 10 by error
print("Parameter Trends for Worst Error:")
print(worst_error[['n','population_size', 'mutation_rate', 'crossover_rate', 'tournament_size','num_elites','crossover_type', 'num_generations']].value_counts())

# Most common parameter values for the top 10 by runtime
print("Parameter Trends for Worst Runtime:")
print(worst_runtime[['n','population_size', 'mutation_rate', 'crossover_rate', 'tournament_size','num_elites','crossover_type', 'num_generations']].value_counts())
```

```python
# Calculate mean parameter values for each group
groups_summary = pd.DataFrame({
    'Best Error': best_error.mean(),
    'Worst Error': worst_error.mean(),
    'Best Runtime': best_runtime.mean(),
    'Worst Runtime': worst_runtime.mean()
})

# Heatmap to visualize differences
sns.heatmap(groups_summary, annot=True, cmap='coolwarm')
plt.title('Parameter Averages: Best and Worst Groups')
plt.show()
```
